```python
gen_count={}
def count_genres(g_list):
    for g in g_list:
        key = g.strip()
        if key not in gen_count:
            gen_count[key] = 1
        else:
             gen_count[key] = gen_count.get(key) + 1
```

```python
df4.GENRE.str.split(",").apply(count_genres)
```

```python
gens = pd.DataFrame([gen_count.keys(), gen_count.values()]).T
```

```python
gens.columns = ["genre", "counts"]
```

```python
gens = gens.sort_values(by= "counts", ascending=False)
```
